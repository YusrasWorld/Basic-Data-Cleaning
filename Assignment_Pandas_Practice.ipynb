```python
# Import the car sales DataFrame with missing data ("../data/car-sales-missing-data.csv")

Missing_Car_Sales = pd.read_csv('Car_Sales_Missing.csv')


# Check out the new DataFrame
Missing_Car_Sales
```

```python
# Fill the Odometer column missing values with the mean of the column inplace

Missing_Car_Sales['Odometer (KM)'].fillna(Missing_Car_Sales['Odometer (KM)'].mean(),inplace = True)
Missing_Car_Sales
```

```python
# Remove the rest of the missing data inplace
Missing_Car_Sales.dropna(inplace = True)
Missing_Car_Sales
```

```python
# Create a "Seats" column where every row has a value of 5

Missing_Car_Sales['Seats'] = 5
Missing_Car_Sales
```

```python
# Create a column called "Engine Size" with random values between 1.3 and 4.5
# Remember: If you're doing it from a Python list, the list has to be the same length
# as the DataFrame

Missing_Car_Sales['Engine Siz'] = ([1.3,1.4,1.6,2.4,3.7,3.9,4.5])
Missing_Car_Sales
```

```python
# Create a column which represents the price of a car per kilometer
# Then view the DataFrame

Missing_Car_Sales["Price"] = Missing_Car_Sales["Price"].astype(float).astype(int)
Missing_Car_Sales["Price per KM"] = Missing_Car_Sales["Price"] / Missing_Car_Sales["Odometer (KM)"]
Missing_Car_Sales
```

```python
# Remove the last column you added using .drop()

Missing_Car_Sales = Missing_Car_Sales.drop("Price per KM", axis=1)
Missing_Car_Sales
```

```python
# Shuffle the DataFrame using sample() with the frac parameter set to 1
# Save the the shuffled DataFrame to a new variable

Missing_Car_Sales_sample = Missing_Car_Sales.sample(frac=1)
Missing_Car_Sales_sample
```

```python
# Reset the indexes of the shuffled DataFrame
Missing_Car_Sales_sample.reset_index()
```

```python
# Change the Odometer values from kilometers to miles using a Lambda function
# Then view the DataFrame

Missing_Car_Sales['Odometer (KM)'] = Missing_Car_Sales['Odometer (KM)'].apply(lambda x : x / 0.6)
Missing_Car_Sales
```

```python
# Change the title of the Odometer (KM) to represent miles instead of kilometers

Missing_Car_Sales = Missing_Car_Sales.rename(columns = {'Odometer (KM)':'Odometer (Miles)'})
Missing_Car_Sales
```
